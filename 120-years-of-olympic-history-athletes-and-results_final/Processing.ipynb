```python
import csv

Database = []
with open("athlete_events.csv", 'r') as csv_file:
    reader = csv.reader(csv_file, delimiter=',')
    for row in reader:
        Database.append(row)

Database.pop(0)

identifier = []
new = [["Country","Year","Season","Bronze","Silver","Gold","Total points"]]


for row in Database:
    country = row[7]
    season = row[10]
    year = row[9]
    if ([country,year,season]) not in identifier:
        new.append([country,year,season,0,0,0,0])
        identifier.append([country,year,season])

print ("pass")
count = 0
for row in Database:
    count += 1;
    if (count%1000==0):
        print(count)
    country = row[7]
    year = row[9]
    season = row[10]
    for newrow in new:
        if ([country,year,season]) == [newrow[0],newrow[1],newrow[2]]:
            if row[14] == "Bronze":
                newrow[3]+=1
                newrow[6]+=1
            elif row[14] == "Silver":
                newrow[4]+=1
                newrow[6]+=3
            elif row[14] == "Gold":
                newrow[5]+=1
                newrow[6]+=5

with open('newfile_summer.csv', mode='w', newline='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerow(new[0])
    for row in new:
        season = row[2]
        if season == "Summer":
            writer.writerow(row)
                      
with open('newfile_winter.csv', mode='w', newline='') as file:
    writer = csv.writer(file, delimiter=',')
    writer.writerow(new[0])
    for row in new:
        season = row[2]
        if season == "Winter":
            writer.writerow(row)
```
