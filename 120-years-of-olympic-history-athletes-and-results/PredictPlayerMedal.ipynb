```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import re
sb.set
```

```python
OlyData = pd.read_csv("athlete_events.csv")
OlyData.head()
df = pd.DataFrame(OlyData['Medal'])
df = df.fillna(0)
df = df.replace('Bronze', 1)
df = df.replace('Silver', 2)
df = df.replace('Gold', 3)
```

```python
df.head
```

```python
OlyDataCat = pd.DataFrame(OlyData[['Sex','Sport','NOC','Medal']])
OlyDataNum = pd.DataFrame(OlyData[['Age','Height','Weight']])
OlyDataNum = OlyDataNum.fillna(0)
OlyDataNum['Medal'] = df
```

```python
#Printing graphs of multiple variables
f, axes = plt.subplots(4, 3, figsize=(18, 20)) #row num/variable num, 3 graphs each
colors = ["r", "g", "b", "c"]

count = 0
for var in OlyDataNum:
    sb.boxplot(OlyDataNum[var], orient = "h", color = colors[count], ax = axes[count,0])
    sb.distplot(OlyDataNum[var], color = colors[count], ax = axes[count,1])
    sb.violinplot(OlyDataNum[var], color = colors[count], ax = axes[count,2])
    count += 1
```

```python
print(OlyDataNum.corr())
f, axes = plt.subplots(1, 1, figsize=(20, 20))
sb.heatmap(OlyDataNum.corr(), vmin = -1, vmax = 1, linewidths = 1,
           annot = True, fmt = ".2f", annot_kws = {"size": 18}, cmap = "RdBu")
```
